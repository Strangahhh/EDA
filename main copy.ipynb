```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import csv

role = []
company = []
date_post = []
highlight_job = []
career_level = []
qualification = []
exp = []
job_type = []
job_description = []
for datalink in datalinks:

    driver = webdriver.Firefox()

    driver.get(f'{datalink}')

    wait = WebDriverWait(driver, 10)

    original_window = driver.current_window_handle

    assert len(driver.window_handles) == 1

    driver.execute_script("document.body.style.transform='scale(0.1)';")
    driver.execute_script("document.body.style.transformOrigin='0 0';")

    time.sleep(2)

    data = driver.page_source
    soup = bs4.BeautifulSoup(data)
    
    time.sleep(1)

    #Role
    class_name = "z1s6m00 _1hbhsw64y y44q7i0 y44q7il _1d0g9qk4 y44q7is y44q7i21"
    h1 = soup.find("h1", class_=class_name)

    if h1:
        role1 = h1.get_text()
        print("role :", role1)
    else:
        print("-")    
    role.append(role1)
    #Company
    span = soup.find("span", class_="z1s6m00 _1hbhsw64y y44q7i0 y44q7i2 y44q7i21 _1d0g9qk4 y44q7ia")

    if span:
        company1 = span.get_text()
        print("Company :", company1)
    else:
        print("-")
    company.append(company1)

    #DATETIME
    span = soup.find_all("span", class_="z1s6m00 _1hbhsw64y y44q7i0 y44q7i1 y44q7i21 y44q7ii")

    if span:
        datetime = span[4].get_text()
        print("Date-post :", datetime)
    else:
        print("-")
    date_post.append(datetime)

    #highlight
    div = soup.find_all("div", class_="z1s6m00 _1hbhsw6r _1hbhsw6p _1hbhsw6a2")
    divselect = div[2:5]
    hl = [div.get_text(strip=True) for div in divselect]
    for h in hl:
        print(f'{h}')
    highlight_job.append(h)
    
    #career
    span1 = soup.find_all("span", class_="z1s6m00 _1hbhsw64y y44q7i0 y44q7i1 y44q7i21 _1d0g9qk4 y44q7ia")
    span1select = span1[15:19]
    textspan1 = [span1.get_text(strip=True) for span1 in span1select]
    for textspan in textspan1:
        print(f'{textspan}')
    career_level.append(textspan1[0])
    qualification.append(textspan1[1])
    exp.append(textspan1[2])
    job_type.append(textspan1[3])

    #JobQ
    div1 = soup.find_all("div", class_="z1s6m00 _5135ge0 _5135ge2")
    jobQ = div1[1].get_text()
    
    sentences = re.split(r'[,.:;]', jobQ)

    for sentence in sentences:
        print(sentence.strip())
    job_description.append(sentence)
    driver.quit()
    
        
```

```python
df = pd.DataFrame()
df['RoleName'] = role
df['Company'] = company
df['Date_post'] = date_post
df['Highlight_job'] = highlight_job
df['Career_level'] = career_level
df['Qualification'] = qualification
df['Exp'] = exp
df['Job_type'] = job_type 
```

```python
import spacy

# Load the English tokenizer, tagger, parser, NER, and word vectors
nlp = spacy.load("en_core_web_sm")

# Your text goes here
text = '''Your long text with sentences, that may need, further splitting.'''

# Process the text
doc = nlp(jobQ)

# Define a maximum sentence length
max_length = 100  # You can adjust this value

# Store the final sentences here
final_sentences = []

# Process sentences from spaCy
for sent in doc.sents:
    # If the sentence is longer than max_length, we might want to split it further
    if len(sent.text) > max_length:
        # Further split the sentence by commas
        sub_sentences = sent.text.split(',')
        # Add non-empty sub sentences to the final list
        final_sentences.extend([sub.strip() for sub in sub_sentences if sub.strip()])
    else:
        # If sentence is shorter than max_length, add it directly
        final_sentences.append(sent.text)

# Now print the final sentences
for sentence in final_sentences:
    print(sentence)
```
